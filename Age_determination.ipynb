```python
def find_age(cluster_ID,Age,mu,ext):
    cluster_name="ngc_"+str(cluster_ID)
    c_name="ngc"+str(cluster_ID)
    
    c_file_name = "Data/"+c_name+".dat"
    print(c_file_name)
    Color,App = np.loadtxt(c_file_name,unpack = True)
    
    
    Chi_square = []
    plt.figure(figsize=[10,10])
    
    

    for a in Age:
        filename="Data/Isochrones/"+c_name+"/a"+str(int(a*1000))+cluster_name+".txt"
        #print(filename)
        f606w,f814w=np.loadtxt(filename,unpack=True,usecols=(12,16))
        mV_fit=convert_to_m(f606w,mu)-ext
    
        for mapp in App:
            i = find_nearest(mV_fit,mapp)
            
            if(i<np.size(Color)):
                color_iso = f606w[i]-f814w[i]
                chi = (color_iso - Color[i])**2
    
        Chi_square.append([a,np.sum(chi)])  
        plt.plot(f606w-f814w,mV_fit-ext,label=str(a)+" Gyr")
    
    plt.gca().invert_yaxis()
    plt.scatter(Color,App,s=1.6,c="grey",alpha=0.2)
    #plt.xlim(-0.5,2)
    plt.xlabel("Color")
    plt.ylabel("Apparent Magnitude")
    plt.title("NGC "+str(cluster_ID))
    plt.legend(loc=3, prop={'size': 12})
    plt.savefig("Figures/Age/"+cluster_name+".pdf")
              
    Chi_square = np.array(Chi_square)
    print(Chi_square)
    print(Chi_square[:,1])
    return Chi_square[np.argmin(Chi_square[:,1])]          
        
              
    
    
```
