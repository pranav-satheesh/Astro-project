```python
mu = 13.94
Mv_NGC104 = mv_NGC104 - mu
Mv_fit = mv_fit - mu
```

```python
def convert_to_m(M,d):
    m = M - 5 + 5 * np.log10(d)
    return m

def find_nearest(array, value):
    array = np.asarray(array)
    idx = (np.abs(array - value)).argmin()
    return idx

def distance(BV_value,BV_standard,Mv_standard,color,mv_c):
    
    i1 = find_nearest(BV_standard,BV_value)
    i2 = find_nearest(color,BV_value)
    
    mu = mv_c[i2] - Mv_standard[i1]
    d = np.power(10, (mu+5)/5)
    return d
```

```python
def find_distance1(cluster_name):
    
    file_name = "Data/"+cluster_name+".dat"
    Color,Abs = np.loadtxt(file_name,unpack = True)
    
    D = []
    
    for BVs in Color:
        D.append(distance(BVs,BV_NGC104,Mv_NGC104,Color,Abs))
        
    n, bins, patches = plt.hist(D,100)
    mode_index = n.argmax()
    dis = (bins[mode_index] + bins[mode_index+1])/2
    print('the distance to the cluster:'+ str(dis))
    
    mV_fit=convert_to_m(Mv_NGC104,dis)
    
    plt.figure()
    plt.scatter(BV_NGC104,mV_fit,s=1.0,c="red",label="standard CMD fit")
    plt.scatter(Color,Abs,s=1.2,c="blue",label= cluster_name)
    #plt.scatter(BV,mV,s=1.2,c="red",alpha=0.1,label="NGC104")
    #plt.plot(Bv_fit,S_mV_fit,"k-")
    plt.axis([-0.5,2.4,25,14])
    plt.xlabel("B-V")
    plt.ylabel("$M_{v}$")
    plt.legend()
    plt.title("d = ")
    
```
