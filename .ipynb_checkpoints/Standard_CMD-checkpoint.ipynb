```python
def convert_to_m(M,d):
    m = M - 5 + 5 * np.log10(d)
    return m
```

```python
def find_nearest(array, value):
    array = np.asarray(array)
    idx = (np.abs(array - value)).argmin()
    return idx
```

```python
def distance(BV_value):
    
    i1 = find_nearest(Bv_fit,BV_value)
    i2 = find_nearest(BV1,BV_value)
    
    mu = mV1[i2] - Mv_fit[i1]
    d = np.power(10, (mu+5)/5)
    return d
```

```python
D = []

for BVs in Bv_fit:
    D.append(distance(BVs))
```

```python
D = np.array(D)
np.argwhere(D<8000)
```

```python
n, bins, patches = plt.hist(D)
```

```python
mode_index = n.argmax()
```

```python
print('the mode:'+ str((bins[mode_index] + bins[mode_index+1])/2))
```
