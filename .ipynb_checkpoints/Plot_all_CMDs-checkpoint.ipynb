```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
plt.style.use('default')
#plt.rcParams['text.usetex'] = True
```

```python
f606w,f814w = np.loadtxt('ngc104.dat',usecols=(7,8),unpack=True)
```

```python
BV = f606w-f814w
```

```python
np.argwhere(BV > 100)
```

```python
BV[378]
```

```python
BV = np.delete(BV,378)
mV = np.delete(f606w,378)
```

```python
f606w[378]
```

```python
f814w[378]
```

```python
np.savetxt('Data/ngc104.dat',np.c_[BV,mV])  
```

```python
plt.scatter(BV,mV,s=1.2,c="purple",alpha=0.3)
plt.axis([-0.2,1.5,25,13])
plt.title("NGC 104")
#plt.grid()
plt.xlabel("$f606w - f814w$")
plt.ylabel("$f606w$")
```
