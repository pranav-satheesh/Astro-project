```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy import stats as st
plt.style.use('default')
```

```python
cluster_name = "ngc4833"
```

```python
file_name = "Data/"+cluster_name+".dat"
Color,Abs = np.loadtxt(file_name,unpack = True)
```

```python
plt.scatter(Color,Abs)
```

```python
idx = np.argwhere(Color <-2)
```

```python
idx
```

```python
Color = np.delete(Color,idx)
Abs = np.delete(Abs,idx)
```

```python
plt.scatter(Color,Abs)
```
