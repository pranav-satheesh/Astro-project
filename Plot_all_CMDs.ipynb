```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
plt.style.use('default')
#plt.rcParams['text.usetex'] = True
```

```python
cluster_name = "ngc6121"
```

```python
f606w,f814w = np.loadtxt('HST_data/'+cluster_name+'.dat',usecols=(7,8),unpack=True)
```

```python
BV = f606w-f814w
mV = f606w
```

```python
plt.scatter(BV,mV)
```

```python
idx1 = np.argwhere(BV>100)
idx2 = np.argwhere(BV<-100)
```

```python
idx2
```

```python
for i1 in idx1:
    print(i1)
    BV=np.delete(BV,i1)
    mV=np.delete(mV,i1)
    
for i2 in idx2:
    #print(i2)
    BV=np.delete(BV,i2)
    mV=np.delete(mV,i2)
```

```python
plt.scatter(BV,mV)
```

```python
plt.scatter(BV,mV,s=1.2,c="purple",alpha=0.3)
plt.axis([-0.2,1.5,25,13])
plt.title(cluster_name)
#plt.grid()
plt.xlabel("$f606w - f814w$")
plt.ylabel("$f606w$")
plt.savefig("Figures/HR/"+cluster_name+".pdf")
```

```python
np.savetxt('Data/'+cluster_name+'.dat',np.c_[BV,mV])  
```

```python
plt.style.use('default')
plt.rc('font', family='serif')
plt.rc('xtick', labelsize='x-small')
plt.rc('ytick', labelsize='x-small')
plt.rcParams['font.size'] = 18
```
